```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("winequalityN.csv")
```

```python
df.tail()
```

```python
x=df.type
y=df.quality
plt.scatter(x,y)
plt.title("WINE QUALITY")
plt.xlabel("type")
plt.ylabel("quality")
plt.legend()
plt.show()
```

```python
plt.bar(x,y,label="BARCHART",color="orange")
plt.title("WINE QUALITY")
plt.xlabel("type")
plt.ylabel("quality")
plt.legend()
plt.show()
```

```python
plt.stem(x,y)
plt.title("WINE QUALITY")
plt.show()
```

```python
plt.plot(x,y)
plt.title("WINE QUALITY")
plt.xlabel("type")
plt.ylabel("quality")
plt.legend()
plt.show()
```

```python
z=df.density
plt.pie(z,y)
plt.title("WINE QUALITY")
plt.show()
```

```python
DS=df.alcohol
plt.hist(DS,color='brown')
plt.title("WINE QUALITY")
plt.show()
```
